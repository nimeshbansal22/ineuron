```python
import os

def replaceText(x,y):
    try:
        f = open("file.txt", "w")
        f1 = open("new.txt","w")
        try:
            f.write(y)
            f1.write(x)
            os.remove("file.txt")
            os.rename("new.txt", "file.txt")
        except Exception as ex:
            print(ex)
        finally:
            f1.close()
        print("File replaced")
    except Exception as ex:
        print(ex)
    
    
    
y = input("Enter the text you want to input into the file")
x = input("Enter text that will replace the existing text:")

replaceText(x,y)
```

```python
ABSTRACT CLASS

from abc import ABC, abstractmethod
 
class Polygon(ABC):
 
    @abstractmethod
    def noofsides(self):
        pass
 
class Triangle(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 3 sides")
 
class Pentagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 5 sides")
 
class Hexagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 6 sides")
 
class Quadrilateral(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 4 sides")
 
# Driver code
R = Triangle()
R.noofsides()
 
K = Quadrilateral()
K.noofsides()
 
R = Pentagon()
R.noofsides()
 
K = Hexagon()
K.noofsides()
```

```python
MULTIPLE INHERITANCE


class Class1:
    def m(self):
        print("In Class1")
       
class Class2(Class1):
    def m(self):
        print("In Class2")
 
class Class3(Class1):
    def m(self):
         print("In Class3")    
     
class Class4(Class2, Class3):
    def m(self):
        print("In Class4")  
 
obj = Class4()
obj.m()
 
Class2.m(obj)
Class3.m(obj)
Class1.m(obj)
```
